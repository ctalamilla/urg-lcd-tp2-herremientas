
## Punto 3

Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron  (columna Survived).

El valor 1 indica que está sobrevivio y 0 no.


```python
survival_counts = df['Survived'].value_counts(normalize=True) * 100


survived_percentage = survival_counts[1]
died_percentage = survival_counts[0]

output_survival = f"Porcentaje de personas que sobrevivieron: {survived_percentage:.2f}%\n"
output_survival += f"Porcentaje de personas que murieron: {died_percentage:.2f}%\n"
```

```python
print(output_survival)
```

```python
labels = ['Sobrevivieron', 'Murieron']
percentages = [survived_percentage, died_percentage]

plt.figure(figsize=(5, 4))
sns.barplot(x=labels, y=percentages)

plt.title("Porcentaje de Supervivientes vs. Fallecidos", fontsize=16)
plt.ylabel("Porcentaje", fontsize=12)

plt.show()
```

## Punto 4

Mostrar por pantalla el porcentaje de personas que sobrevivieron (columna Survived) en cada clase (columna Pclass).

El valor 1 indica que sobrevivio y 0 no.

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100


output_survival_by_class = ""
for pclass, survival_rate in survival_by_class.items():
    output_survival_by_class += f"Clase {pclass}: {survival_rate:.2f}% sobrevivieron\n"
```

```python
print(output_survival_by_class)
```

```python
plt.figure(figsize=(6, 5))
ax = sns.barplot(x=survival_by_class.index, y=survival_by_class.values)

plt.title("Porcentaje de Supervivientes por Clase", fontsize=16)
plt.xlabel("Clase", fontsize=12)
plt.ylabel("Porcentaje de Supervivientes", fontsize=12)

# Add labels on each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')

plt.show()
```

## Punto 2

Crear un columna `person` y completarla con la búsqueda de la  entidad persona dentro de la columna `news`. La entidad persona es un nombre y apellido.

Tiene que ser una aproximación. No va a ser perfecta por falta de contexto.

```python
import re
```

```python
# Definir un patrón básico para nombres y apellidos (con mayúscula inicial)
patron_nombres = r'\b[A-ZÁÉÍÓÚÑ][a-záéíóúñ]+(?:\s[A-ZÁÉÍÓÚÑ][a-záéíóúñ]+)+'  # Patrones para nombres compuestos

def extract_person(text):
    possible_names = re.findall(patron_nombres, text)
    return possible_names
```

```python
df['person'] = df['news'].apply(extract_person)
```

```python
df.sample(5)
```

## Punto 3

Crear un columna `domain` y completarla con el dominio del sitio dentro de la columna `url`. Un dominio es `larepublica.co`.

```python
def extract_domain(url):
    domain_match = re.search(r'https?://(www\.)?([^/]+)', url)
    if domain_match:
        return domain_match.group(2)
    return None
```

```python
df['domain'] = df['url'].apply(extract_domain)
```

```python
df.sample(5)
```

## Punto 4

Crear una columna llamada `words` y contabilizar la cantidad de palabras que hay en la columna `news`.

```python
def count_words(text):
    return len(text.split())
```

```python
df['words'] = df['news'].apply(count_words)
```
